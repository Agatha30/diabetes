# **Explanation of data, data source and features**

1. **Title of data:** Women Diabetes dataset
2. **Source:** Bulawayo Central Hospital
3. **Publisher:** Dr Mbonisi Nyathi
4. **Description:** The dataset is a collection of data females suspected of being diabetics who were admitted at the hospital.
5. **Number of records:** 800
6. **Number of data features:** 9
7. **Feature description:**
>*   Number of times pregnant
>*   Plasma glucose concentration a 2 hours in an oral glucose tolerance test
>*   Diastolic blood pressure (mm Hg)
>*   Triceps skin fold thickness (mm)
>*   2-Hour serum insulin (mu U/ml)
>*   Body mass index (weight in kg/(height in m)^2)
>*   Diabetes pedigree function
>*   Age (years)
>*   Class variable (0 or 1)
8. **Class intepretation:** Class value 1 means "Tested Positive for diabetes"
9. **Objective of the Data Analytics:** The objective in this project is to train a model that *binary classifies* patients into diabetic or non diabetic.














We will use Keras as the deep learning framework. We apply the multi-perceptron feedfoward recurrent neural network architecture called a **sequential model**

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

Load the dataset from the github repo of the researcher and split the data into features X and classes y

```python
dataset = loadtxt('https://raw.githubusercontent.com/mpumzee/diabetics-data/master/byo-central-hospital-diabetes.data.csv', delimiter=',')
X = dataset[:,0:8]
y = dataset[:,8]
```

Split the data into training and testing at a ration of 67% training and 33% testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

We define the keras model. The model will have 8 feature inputs and 3 layers. The first layer which is the input layer will have 12 nodes which is a random chosen number that can be changed during optimization. We chose the ReLU activation function because ReLU function is far more computationally efficient when compared to the sigmoid and tanh function seeing it does not activate all the neurons at the same time.



```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

# Compiling the keras model
**Binary loss function:** Since our model is a binary classification model we therefore use the binary loss function as a compilation parameter.

**Adam Optimizer**
This is a popular version of gradient descent because it automatically tunes itself and gives good results in a wide range of problems


```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# Fitting the keras model on the dataset
Here we traing the model and in the process demonstrate the change in accuracy from around 65% to 75% when we change the numbe of epochs from 150 to 200 and the batch size from 10 to 15

```python
model.fit(X_train, y_train, epochs=200, batch_size=15)
```

# Model Evaluation
We will use Accuracy as a means of model evaluation. Accuracy is a **good perfomance metric** because the target variable classes in the data are nearly balanced.

```python
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```

**Predictions**: This is the actual usage of the model and therefore an outcome of the project

```python
predictions = model.predict(X_test)
```

Here we make predictions from the test data

```python
predictions = (model.predict(X) > 0.5).astype(int)
```
